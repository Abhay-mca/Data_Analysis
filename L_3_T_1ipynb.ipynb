```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
url = "https://raw.githubusercontent.com/Abhay-mca/Data_Analysis/main/Dataset%20.csv"
df = pd.read_csv(url)
```

```python
df = df.dropna(subset=['Rating text'])
```

```python
words = df['Rating text'].str.lower().str.split()
all_words = [word for sublist in words for word in sublist]
word_counts = Counter(all_words)
```

```python
print("Most common words in Rating text:")
print(word_counts.most_common(10))
```

```python
plt.figure(figsize=(8,5))
df['Rating text'].value_counts().plot(kind='bar', color='teal', edgecolor='black')
plt.title("Distribution of Rating Text")
plt.xlabel("Rating Text")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
