```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/Abhay-mca/Data_Analysis/main/Dataset%20.csv"
df = pd.read_csv(url)
```

```python
print(df['Price range'].value_counts())
```

```python
price_range_counts = df['Price range'].value_counts()
total_restaurants = df.shape[0]
price_range_percentages = (price_range_counts / total_restaurants) * 100
price_range_percentages = price_range_percentages.round(2)
```

```python
print("\nPercentage of restaurants in each price range:")
print(price_range_percentages)
```

```python
plt.figure(figsize=(8, 5))
price_range_percentages.sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Distribution of Price Ranges Among Restaurants')
plt.xlabel('Price Range (1 = Low, 4 = High)')
plt.ylabel('Percentage of Restaurants')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
